```python
#Here we make analysis on NBA data
#Data can be found from here: http://51.91.120.89/extras/nba.csv
import pandas as pd
```

```python
url = "http://51.91.120.89/extras/nba.csv"
df = pd.read_csv(url, index_col='Name')
df
```

```python
#Let's get rid of NAs
df.dropna(axis=0, how='all', inplace=True)
df
```

```python
df.isna().sum()
```

```python
#Let's change NAs in College to empty strings
df["College"].replace(pd.np.NaN, "", inplace=True)
df.isna().sum()
```

```python
#Since the weights are obviously in pounds, let's convert them to kilograms
# 1 pound = 2.2 kg
df["Weight"] /= 2.2
df
```

```python
#Since the heights are obviously in inches, let's convert them to centimetres
# 1 inch = 2.54 centimetres

#To do that let us create a sub- function to make the changes
def convert_height(s):
    feet, inches = s.split("-")
    return int(feet)*30.4+int(inches)*2.54
```

```python
#Let's see if the code worked:

convert_height("6-10")
```

```python
#Let's make the changes in the data
df["Height"] = df["Height"].apply(convert_height)
```

```python
#let's see if it worked: Yes
df
```

```python
#Let's sort by age
df.sort_values("Age")
```

```python
#Let's sort by age (from max to min)
df.sort_values("Age", ascending=False)
```

```python
#Let's sort by youngest age and tallest height
df.sort_values(["Age","Height"], ascending=[False, True]).head(10)
```

```python
#Let's sort by top 10 that has shortest height
df.nsmallest(10,"Height")
```

```python
#Let's sort by top 10 that is oldest
df.nlargest(10,"Age")
```

```python
#Let's see the players that are higher than 2 meters
df[(df.Height>200)]
```
